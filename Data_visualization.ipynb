## Department wise performances

```python
df.groupby(['EmpDepartment', 'PerformanceRating'])['Gender'].value_counts()
```

```python
# Create a figure and axes
fig, axes = plt.subplots(2, 2, figsize=(12,12))

# Plotting Countplot of Performance Rating by Department
sns.countplot(x=df['EmpDepartment'], hue=df['PerformanceRating'], ax=axes[0, 0])
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[0, 0].set_title('Countplot of Performance Rating by Department')

# Plotting the violinplot in the second subplot
sns.violinplot(x=df['EmpDepartment'], y=df['PerformanceRating'], hue=df['Gender'], split=True, ax=axes[0, 1])
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[0, 1].set_title('Violinplot of Performance Rating by Department and Gender')

# Plotting the violinplot with age and Emp department
sns.violinplot(x=df['EmpDepartment'], y=df['Age'], hue=df['Gender'], split=True, ax=axes[1, 0])
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[1, 0].set_title('Violinplot of Age with Empdepartment')

# Plotting violinplot of work experience and Emp Department
sns.violinplot(x=df['EmpDepartment'], y=df['TotalWorkExperienceInYears'], hue=df['Gender'], split=True, ax=axes[1, 1])
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[1, 1].set_title('Violinplot of Work Experience with Emp Department')

plt.suptitle('Employee Department vs Performance Rating', fontsize=16)
plt.tight_layout()


plt.show()
```

### Insights
1. The majority of employees in the **Sales department** achieve level 3 performance ratings, with males slightly outperforming females, while work experience shows minimal impact on performance.
2. Most employees in **Human Resources** attain level 3 performance ratings, with older employees performing less effectively, although female employees excel, and work experience significantly influences performance.
3. The **Development department** sees a large number of employees achieving level 3 performance across all age groups, with a balanced performance distribution between genders. 
4. **Data Science** records the highest average of level 3 performance, with fewer employees at level 2, and age having little impact on performance, while males tend to perform well, and work experience is influential.
   
5. Age does not significantly influence performance in the **Research & Development department**, which showcases strong female performers. 
6. Performance in the **Finance department** declines with age, yet males perform well, and experience inversely affects performance levels.

# Top 3 Important Factors effecting employee performance 

```python
X = df.drop(['PerformanceRating','EmpNumber' ], axis=1)
y = df['PerformanceRating']
```

```python
X = pd.get_dummies(X, drop_first=True, dtype='int')
```

```python
rf_model = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_model.fit(X, y)
```

```python
# Feature importance analysis
feature_importances =rf_model.feature_importances_
feature_names = X.columns
feature_imp = pd.Series(feature_importances)
feature_imp.index = feature_names
feature_imp = feature_imp.sort_values(ascending=False)
```

```python
feature_imp.head(3).plot(kind='barh', legend=False)
```
