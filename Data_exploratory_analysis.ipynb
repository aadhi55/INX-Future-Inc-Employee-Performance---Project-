# Exploratory Data Analysis

```python
from tabulate import tabulate
for col in df.columns[1:]:
    if df[col].dtype=='object':
        print(col)
        print(tabulate(df[col].value_counts().reset_index(), headers = ['Value', 'Count'], tablefmt = 'grid'))
```

### Countplot

```python
catcols = [col  for col in df.columns[1:] if df[col].dtype == 'object']
```

```python
# Finding the value_counts using countplot
plt.figure(figsize=(15, 8), facecolor='white') 
plotnumber = 1
for col in catcols:
    ax = plt.subplot(2, 4, plotnumber)  
    sns.countplot(x=df[col], ax=ax)
    plt.xticks(rotation=45, fontsize=10, ha='right')
    plt.xlabel(col)
    plotnumber += 1

plt.tight_layout()
plt.show()
```

### Insights from Exploratory Data Analysis (EDA)
1. **Gender Distribution**:
   - There are more male employees (725) compared to female employees (475).

2. **Education Background**:
   - The most common educational backgrounds among employees are Life Sciences (492) and Medical (384), followed by Marketing (137) and Technical Degree (100).

3. **Marital Status**:
   - Majority of the employees are married (548), followed by single (384) and divorced (268).

4. **Employee Department**:
   - The largest departments in terms of employee count are Sales (373), Development (361), and Research & Development (343), while departments like Data Science (20) and Finance (49) have fewer employees.

5. **Employee Job Role**:
   - Common job roles include Sales Executive (270), Developer (236), Manager R&D (94), and Research Scientist (77).

6. **Business Travel Frequency**:
   - Most employees travel rarely for business purposes (846), followed by frequent travelers (222), and those who do not travel at all (132).

7. **Overtime**:
   - A majority of employees (847) do not work overtime, while a smaller proportion (353) do.

8. **Attrition**:
   - The attrition rate is relatively low, with 1022 employees staying in the company (No) and 178 employees leaving (Yes).

### Distribution of numerical data

```python
numcols = [col for col in df.columns if df[col].dtype!='object']
```

```python
fig = plt.figure(figsize = (18,12), facecolor='white')
for i, col in enumerate(numcols[:-1], start = 1):
    ax = fig.add_subplot(5,4,i)
    sns.histplot(x = df[col],kde = True, ax = ax)
     
plt.tight_layout()
plt.show()
```

### Insights
